```python
#ok i figured out that i have to first split the values and make them into single values first.
mdata_split_genre = mdata.copy()
#let me just copy the contents into a different variable,so no mixup or change of data happens
#so first to get the whole splitted data of genres,.split(|),splits the whole thing wherever there is |
#apply(pd.Series, 1) is used for puttin in all the values , 1 is for true,meaning to be done.
#stack() puts in all the values one after the other in some order.
#reset_index(level=1, drop=True) will tell to add a total different array or a set,level is basically to remove the first level 
#and drop is not to add it into the dataset as another column.
#split_genre = mdata_split_genre['genres'].str.split('|').apply(pd.Series, 1).stack().reset_index(level=1, drop=True)
#now ill just drop the genres column and will specify that not to make changes in the original column instead put it as diff col
#and then will join the splitted genre,where we split it with |


split_genre = mdata_split_genre['genres'].str.split('|').apply(pd.Series, 1).stack().reset_index(level=1, drop=True)
split_genre.name = 'genre_split'
mdata_split_genre = mdata_split_genre.drop(['genres'], axis=1).join(split_genre)
```
